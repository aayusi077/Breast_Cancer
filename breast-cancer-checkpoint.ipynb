```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import datetime
from pandas.plotting import andrews_curves, parallel_coordinates, radviz

from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score 
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import VotingClassifier
```

```python
df = pd.read_csv('breast-cancer.csv')
print(df.shape)
df.head()
```

```python
def prepare(df):
    # benign is good
    df.diagnosis = df.diagnosis.map({'M':1, 'B':0})
    
    # drop columns as their correlation is quite low
    df.drop(['fractal_dimension_mean', 'id', 'texture_se', 'symmetry_se'], axis=1, inplace=True)
    

    return df
```

```python
df = prepare(df)
df.head()
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:,0]

ss=StandardScaler()

X_train,X_val,y_train,y_val= train_test_split(X,y,test_size=0.2)
X_train=ss.fit_transform(X_train)
X_val=ss.transform(X_val)
```

```python
def learn(X_train, X_val, y_train,y_val, clf):
    #cv = cross_val_score(clf,(X_train, y_train, cv=5 )
    clf.fit(X_train, y_train)
    train_clf_pred=clf.predict(X_train)
    #print(classification_report(y_train,train_clf_pred))
    #print(confusion_matrix(y_train,train_clf_pred))

    print('_______________ EVALUATION OF PREDICTION _______________')
    test_clf_pred =clf.predict(X_val)
    print(classification_report(y_val,test_clf_pred))
    print(confusion_matrix(y_val,test_clf_pred))

    acc_lr = accuracy_score(y_val, test_clf_pred)
    print (acc_lr)
    return accuracy_score(y_val, test_clf_pred)
```

```python
models = pd.DataFrame(columns=['KNN', 'DecisionTree', 'RandomForest', 'SVM','LogiRegr',
                               'MultyLayerPerc', 'Voting'])
models
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
models.loc[1,'DecisionTree']=  learn(X_train,X_val,y_train,y_val, clf)
```

K-nearest Neighbours

```python
knn = KNeighborsClassifier(n_neighbors = 4)
models.loc[1,'KNN'] = learn(X_train,X_val,y_train,y_val, knn)
```

Random Forest

```python
model1=RandomForestClassifier(n_estimators=600,max_depth=5)
models.loc[1,'RandomForest'] = learn(X_train,X_val,y_train,y_val, model1)
```

Logistic Regression

```python
lr = LogisticRegression(random_state = 0)
models.loc[1,'LogiRegr'] = learn(X_train,X_val,y_train,y_val, lr)
```

SVM

```python
svc = SVC(kernel = 'linear', random_state = 0, probability=True)
models.loc[1,'SVM'] = learn(X_train,X_val,y_train,y_val, svc)
```

Multilayer Perceptron

```python
mlp = MLPClassifier(alpha=.5, max_iter=2000)
models.loc[1,'MultyLayerPerc'] = learn(X_train,X_val,y_train,y_val, mlp)
```

Voting Classifier

```python
voting_clf = VotingClassifier(estimators=[('lr', lr),('knn',knn),('clf', clf),
                                         ('RandomForest', model1), ('svc', svc),
                                         ('mlp', mlp)], voting='soft')

cv = cross_val_score(voting_clf, X_train, y_train, cv=5)

models.loc[1,'Voting'] = learn(X_train,X_val,y_train,y_val, voting_clf).mean()
```
